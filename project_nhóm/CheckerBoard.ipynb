Tạo bàn cờ trắng đen sử dụng thư viện openCV và nupmy


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Kích thước mỗi ô vuông trên bàn cờ
square_size = 50

# Số lượng ô vuông trên bàn cờ
num_squares = 8

# Tạo một ma trận 2D với các ô trắng và đen xen kẽ
chessboard = np.zeros((square_size * num_squares, square_size * num_squares))

for i in range(num_squares):
    for j in range(num_squares):
        if (i + j) % 2 == 0:
            chessboard[i * square_size:(i + 1) * square_size, j * square_size:(j + 1) * square_size] = 255




# Hiển thị hình ảnh bàn cờ
cv2.imshow("Chessboard", chessboard)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
