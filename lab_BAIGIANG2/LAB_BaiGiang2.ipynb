1.1.4 imread() with different color format

* IMREAD_ANYDEPTH=2
* IMREAD_COLOR=1
* IMREAD_GRAYSCALE=0
* IMREAD_LOAD_GDAL=8

CÂU 2: CÁC ẢNH XÁM VỚI 4,2 BIT

```python
img_xam_4bit = (img_GRAYSCALE//16) #giảm xuống 16 mứa xám (4bit)
img_xam_2bit = (img_GRAYSCALE//64) #giảm xuống 4 mứa xám (2bit)
img_xam_1bit = (img_GRAYSCALE//128) #giảm xuống 2 mứa xám (1bit)
```

```python
#ảnh xám gốc
plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
plt.imshow(img_GRAYSCALE,cmap="gray")
plt.title("ảnh xám gốc")

#ảnh xám 4 bit
plt.subplot(1,4,2)
plt.imshow(img_xam_4bit,cmap="gray")
plt.title("ảnh xám 4 bit")

#ảnh xám 2 bit
plt.subplot(1,4,3)
plt.imshow(img_xam_2bit,cmap="gray")
plt.title("ảnh xám 2 bit")

#ảnh xám 1 bit
plt.subplot(1,4,4)
plt.imshow(img_xam_1bit,cmap="gray")
plt.title("ảnh xám 1 bit")

plt.show()
```
